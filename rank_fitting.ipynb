# Top K operator

```python
class TopK_custom(torch.nn.Module):
    def __init__(self, k, device, epsilon=0.1, max_iter = 200):
        super(TopK_custom, self).__init__()
        self.k = k
        self.epsilon = epsilon
        self.anchors = torch.FloatTensor([i for i in range(k+1)]).view([1,1, k+1]).to(device)
        self.max_iter = max_iter
        self.device = device
        

    def forward(self, scores):
        device = self.device
        
        bs, n = scores.size()
        scores = scores.view([bs, n, 1])
        
        #find the -inf value and replace it with the minimum value except -inf
        scores_ = scores.clone().detach()
        max_scores = torch.max(scores_).detach()
        scores_[scores_==float('-inf')] = float('inf')
        min_scores = torch.min(scores_).detach()
        filled_value = min_scores - (max_scores-min_scores)
        mask = scores==float('-inf')
        scores.masked_fill(mask, filled_value)
        
        C = (scores-self.anchors)**2
        C = C / (C.max().detach())
        
        mu = torch.ones([1, n, 1], requires_grad=False).to(device)
        nu = [1 for _ in range(self.k)]
        nu.append((n-self.k))
        nu = torch.FloatTensor(nu).view([1,1,self.k+1]).to(device)

            
        Gamma = TopKFunc.apply(C, mu, nu, self.epsilon, self.max_iter)
        
        return Gamma.transpose(-1,-2)



class TopKFunc(torch.autograd.Function):
    @staticmethod
    def forward(ctx, C, mu, nu, epsilon, max_iter):
        bs, n, k_ = C.size()
        device = C.get_device()
        with torch.no_grad():
            f = torch.zeros([bs, n, 1]).to(device)
            g = torch.zeros([bs, 1, k_]).to(device)
            Gamma_old = torch.exp((-C+f+g)/epsilon)
            
            def min_epsilon_row(Z, epsilon):
                return -epsilon*torch.logsumexp((-C+f+g)/epsilon, -1, keepdim=True)
            
            def min_epsilon_col(Z, epsilon):
                return -epsilon*torch.logsumexp((-C+f+g)/epsilon, -2, keepdim=True)
    
            for i in range(max_iter):
                
                f = min_epsilon_row(C-f-g, epsilon)+f+epsilon*torch.log(mu)
                g = min_epsilon_col(C-f-g, epsilon)+ g +epsilon*torch.log(nu)
                
                Gamma = torch.exp((-C+f+g)/epsilon)
                
                Gamma_diff = torch.norm((Gamma-Gamma_old) / bs / n/ k_,"fro")
#                 nu_cond = torch.norm((torch.matmul(Gamma.permute(0,2,1),mu[0,:,:]).permute(0,2,1) - nu) / bs,"fro")
#                 mu_cond = torch.norm((torch.matmul(Gamma,nu.view(k_,1)) - mu) / bs,"fro")
                

                if (Gamma_diff < 1e-5):
                    #print("stop at ",i," step")
                    break
                
                Gamma_old = Gamma
 
        
            ctx.save_for_backward(mu, nu, Gamma)
            ctx.epsilon = epsilon      
        return Gamma

    @staticmethod
    def backward(ctx, grad_output_Gamma):
        
        
        epsilon = ctx.epsilon
        mu, nu, Gamma = ctx.saved_tensors
        # mu [1, n, 1]
        # nu [1, 1, k+1]
        #Gamma [bs, n, k+1]
                  
        device = Gamma.get_device()
        with torch.no_grad():
            nu_ = nu[:,:,:-1]
            Gamma_ = Gamma[:,:,:-1]
    
            bs, n, k_ = Gamma.size()
            
            inv_mu = 1./(mu.view([1,-1]))  #[1, n]
            Kappa = torch.diag_embed(nu_.squeeze(-2)) \
                    -torch.matmul(Gamma_.transpose(-1, -2) * inv_mu.unsqueeze(-2), Gamma_)   #[bs, k, k]
            #print(Kappa, Gamma_)
            padding_value = 1e-10
            ridge = torch.ones([bs, k_-1]).diag_embed().to(device)
            inv_Kappa = torch.inverse(Kappa+ridge*padding_value) #[bs, k, k]
            #print(Kappa, inv_Kappa) 
            mu_Gamma_Kappa = (inv_mu.unsqueeze(-1)*Gamma_).matmul(inv_Kappa) #[bs, n, k]
            H1 = inv_mu.diag_embed() + mu_Gamma_Kappa.matmul(Gamma_.transpose(-1, -2))*inv_mu.unsqueeze(-2) #[bs, n, n]
            H2 = - mu_Gamma_Kappa  #[bs, n, k]
            H3 = H2.transpose(-1,-2) #[bs, k, n]
            H4 = inv_Kappa #[bs, k, k]
    
            H2_pad = F.pad(H2, pad=(0, 1), mode='constant', value=0) 
            H4_pad = F.pad(H4, pad=(0, 1), mode='constant', value=0)
            grad_f_C =  H1.unsqueeze(-1)*Gamma.unsqueeze(-3) \
                       + H2_pad.unsqueeze(-2)*Gamma.unsqueeze(-3) #[bs, n, n, k+1]
            grad_g_C =  H3.unsqueeze(-1)*Gamma.unsqueeze(-3) \
                       + H4_pad.unsqueeze(-2)*Gamma.unsqueeze(-3) #[bs, k, n, k+1]
    
            grad_g_C_pad = F.pad(grad_g_C, pad=(0, 0, 0, 0, 0, 1), mode='constant', value=0)
            grad_C1 = grad_output_Gamma * Gamma
            grad_C2 = torch.sum(grad_C1.view([bs, n, k_, 1, 1])*grad_f_C.unsqueeze(-3), dim=(1,2))
            grad_C3 = torch.sum(grad_C1.view([bs, n, k_, 1, 1])*grad_g_C_pad.unsqueeze(-4), dim=(1,2))
    
            grad_C = (-grad_C1+grad_C2+grad_C3)/epsilon
                   
        return grad_C, None, None, None, None
```

# define a simple MLP

```python
class Net_sigmoid(nn.Module):
    
    def __init__(self, dim):
        super(Net_sigmoid, self).__init__()
        # First fully connected layer
        self.dim = dim
        self.fc1 = nn.Linear(dim, 256)
        # Second fully connected layer that outputs our 10 labels
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 1)
        
    def forward(self, x):
        # Pass data through fc1
        bz, qn, dim = x.shape
        out = self.fc1(x.view(-1,dim))
        out = torch.relu(out)
        out = self.fc2(out)
        out = torch.relu(out)
        out = self.fc3(out)

        # Apply softmax to x
        out = torch.sigmoid(out)
        return out.view(bz,-1)
    
class Net(nn.Module):
    
    def __init__(self, dim):
        super(Net, self).__init__()
        # First fully connected layer
        self.dim = dim
        self.fc1 = nn.Linear(dim, 256)
        # Second fully connected layer that outputs our 10 labels
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 1)
        
    def forward(self, x):
        # Pass data through fc1
        bz, qn, dim = x.shape
        out = self.fc1(x.view(-1,dim))
        out = torch.relu(out)
        out = self.fc2(out)
        out = torch.relu(out)
        out = self.fc3(out)

        return out.view(bz,-1)
```

# fit the function with value data

```python
def index_iterator(data_len, batch_size, shuffle=True):
    if shuffle:
        indices = np.random.permutation(data_len)
    else:
        indices = np.arange(data_len)
    for i in range(0, data_len, batch_size):
        yield indices[i:i+batch_size]

def check_value_model(model,f):
    x_list = np.linspace(0,1,100).reshape(-1,1)
    gt_list = f(x_list)
    with torch.no_grad():
        est_list = model(torch.FloatTensor(x_list).unsqueeze(0)).flatten().numpy()
    plt.plot(x_list,gt_list, label = "gt")
    plt.plot(x_list,est_list, label = "est")
    plt.legend()
    plt.show()

def fit_value(model,x_data,f_data, epoch = 10, batchsize = 1, lr = 0.01):
    data_len = x_data.shape[0]
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = torch.nn.MSELoss(reduction='mean')
    for _ in range(epoch):
        for batch_ind in index_iterator(data_len,batchsize):
            optimizer.zero_grad()
            loss = loss_fn(model(x_data[batch_ind]),f_data[batch_ind])
            loss.backward()
            optimizer.step()
```

```python
mlp = Net(dim)
print(mlp)
fit_value(mlp,x_data,f_data)
check_value_model(mlp,f)
```

# fit the function with rank data 

```python
def check_model(model,f,x_test_data,rank_test_data, plot = True):
    device = next(model.parameters()).device

    with torch.no_grad():
        est_test = model(x_test_data.to(device)).cpu().numpy()
    est_rank = np.apply_along_axis(lambda x: np.argsort(x)[:topk],1,est_test)
    
    acc = np.mean(rank_test_data == est_rank) * 100
    
    if plot:
        x_list = np.linspace(0,1,100).reshape(-1,1)
        gt_list = f(x_list)
        
        with torch.no_grad():
            est_list = model(torch.FloatTensor(x_list).unsqueeze(0).to(device)).cpu().flatten().numpy()
        
        
        plt.figure(figsize=(16,8))
        plt.subplot(1,2,1)
        plt.plot(x_list,gt_list, label = "gt")
        plt.legend()
        plt.subplot(1,2,2)
        plt.plot(x_list,est_list, label = "est")
        plt.legend()
        plt.title("Acc: {}%".format(acc))
        plt.show()
    return acc

def fit_rank(model,x_data,perm_data,x_test_data,rank_test_data, epoch = 100, batchsize = 8, lr = 0.001, tol = 5, plot = True):
    data_len = x_data.shape[0]
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = torch.nn.MSELoss(reduction='mean')
    device = x_data.get_device()
    soft_sorted_topk = TopK_custom(perm_data.shape[1] - 1, device, epsilon=0.01, max_iter = 200)
    avg_loss = 0
    total_num = 0 
    
    best = 0
    no_improve_ep = 0
    best_model = None
    for ep in range(epoch):
        no_improve_ep += 1
        for batch_ind in index_iterator(data_len,batchsize):
            optimizer.zero_grad()
            scores = model(x_data[batch_ind])
            perm_mat = soft_sorted_topk(scores)
            loss = loss_fn(perm_mat, perm_data[batch_ind])
            loss.backward()
            optimizer.step()
            avg_loss += loss.item() * len(batch_ind)
            total_num += len(batch_ind)
            
        acc = check_model(model,f,x_test_data,rank_test_data,False)
        
        
        
        print(ep,avg_loss/total_num,acc)
        if ep % 10 == 0:
            check_model(model,f,x_test_data,rank_test_data, plot)
            
        if acc > best:
            best = acc
            no_improve_ep = 0
            best_model = deepcopy(model.state_dict())
            
        if no_improve_ep > tol:
            break
            
    return best_model

def constuct_perm_data(f_data,topk):
    data_len, query_num = f_data.shape
    rank_data = np.apply_along_axis(lambda x: np.argsort(x)[:topk],1,f_data)
    permutation_data = []
    for ii in range(data_len):
        perm_mat = np.zeros((topk + 1, query_num))
        for jj in range(query_num):
            if jj in rank_data[ii,:]:
                now_ind = np.where(rank_data[ii,:] == jj)[0][0]
                perm_mat[now_ind,jj] = 1
            else:
                perm_mat[topk,jj] = 1
        permutation_data.append(perm_mat[np.newaxis,...])
    permutation_data = np.concatenate(permutation_data)

    
    return torch.FloatTensor(permutation_data)
```

```python
topk  = 2
permutation_data = constuct_perm_data(f_data.numpy(),topk)
```

```python
mlp = Net_sigmoid(dim)
print(mlp)
best_state_dict = fit_rank(mlp.cuda(),x_data.cuda(),permutation_data.cuda(),x_test_data,rank_test_data,lr=0.001,batchsize=16)
mlp.load_state_dict(best_state_dict)
check_model(mlp,f,x_test_data,rank_test_data)
```

# method from instruct GPT - shuffle comparison in one dataset

```python
def constuct_comp_data(x_data, f_data, topk):
    data_len, query_num = f_data.shape
    rank_data = np.apply_along_axis(lambda x: np.argsort(x)[:topk],1,f_data)
    shuffled_comp_data_x1 = []
    shuffled_comp_data_x0 = []
    for ii in range(data_len):
        worst_inds = [ind for ind in range(query_num) if ind not in rank_data[ii,:]]
        for jj, sup_ind in enumerate(rank_data[ii,:]):
            if jj < (len(rank_data[ii,:]) - 1):
                for qq, inf_ind in enumerate(rank_data[ii,(jj+1):]):
                    shuffled_comp_data_x0.append(x_data[ii,sup_ind,:])
                    shuffled_comp_data_x1.append(x_data[ii,inf_ind,:])
            for inf_ind in worst_inds:
                shuffled_comp_data_x0.append(x_data[ii,sup_ind,:])
                shuffled_comp_data_x1.append(x_data[ii,inf_ind,:])
    
    
    return torch.FloatTensor(np.array(shuffled_comp_data_x0)), torch.FloatTensor(np.array(shuffled_comp_data_x1))
    
```

```python
topk  = 2
comp_data_x0, comp_data_x1  = constuct_comp_data(x_data.numpy(), f_data.numpy(), topk)
```

```python
def fit_comp(model, comp_data_x0, comp_data_x1,x_test_data,rank_test_data, epoch = 100, batchsize = 8, lr = 0.001, tol = 5, plot = True):
    data_len = comp_data_x0.shape[0]
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    device = comp_data_x0.get_device()
    avg_loss = 0
    total_num = 0 
    
    best = 0
    no_improve_ep = 0
    best_model = None
    for ep in range(epoch):
        no_improve_ep += 1
        for batch_ind in index_iterator(data_len,batchsize):
            optimizer.zero_grad()
            sup_scores = model(comp_data_x0[batch_ind].unsqueeze(0))
            inf_scores = model(comp_data_x1[batch_ind].unsqueeze(0))
            
            loss = torch.mean(torch.log(torch.sigmoid(sup_scores - inf_scores)))
            loss.backward()
            optimizer.step()
            avg_loss += loss.item() * len(batch_ind)
            total_num += len(batch_ind)
        
        acc = check_model(model,f,x_test_data,rank_test_data,False)
        
        
        
        print(ep,avg_loss/total_num,acc)
        if ep % 10 == 0:
            check_model(model,f,x_test_data,rank_test_data, plot)
            
        if acc > best:
            best = acc
            no_improve_ep = 0
            best_model = deepcopy(model.state_dict())
            
        if no_improve_ep > tol:
            break
            
    return best_model
```

```python
mlp = Net_sigmoid(dim)
print(mlp)
best_state_dict = fit_comp(mlp.cuda(), comp_data_x0.cuda(), comp_data_x1.cuda(),x_test_data,rank_test_data, epoch = 100, batchsize = 32, lr = 0.001)
mlp.load_state_dict(best_state_dict)
check_model(mlp,f,x_test_data,rank_test_data)
```

# method from instruct GPT - one rank as a batch

```python
def constuct_batch_comp_data(x_data, f_data, topk):
    data_len, query_num = f_data.shape
    rank_data = np.apply_along_axis(lambda x: np.argsort(x)[:topk],1,f_data)
    total_comp_data = []
    for ii in range(data_len):
        comp_data_sup = []
        comp_data_inf = []
        
        worst_inds = [ind for ind in range(query_num) if ind not in rank_data[ii,:]]
        for jj, sup_ind in enumerate(rank_data[ii,:]):
            if jj < (len(rank_data[ii,:]) - 1):
                for qq, inf_ind in enumerate(rank_data[ii,(jj+1):]):
                    comp_data_sup.append(sup_ind)
                    comp_data_inf.append(inf_ind)
            for inf_ind in worst_inds:
                comp_data_sup.append(sup_ind)
                comp_data_inf.append(inf_ind)
                
        total_comp_data.append((comp_data_sup,comp_data_inf))
    
    
    return np.array(total_comp_data)

def fit_batch_comp(model,x_data,comp_data,x_test_data,rank_test_data, epoch = 100, batchsize = 1, lr = 0.001, tol = 5, plot = True):
    data_len, query_num, dim = x_data.shape
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = torch.nn.MSELoss(reduction='mean')
    device = x_data.get_device()
    avg_loss = 0
    total_num = 0 
    
    best = 0
    no_improve_ep = 0
    best_model = None
    for ep in range(epoch):
        no_improve_ep += 1
        for batch_ind in index_iterator(data_len,batchsize):
            optimizer.zero_grad()
            scores = model(x_data[batch_ind])
            batch_sup = comp_data[batch_ind,0,:]
            batch_inf = comp_data[batch_ind,1,:]
            
            loss = 0
            for ind in range(len(batch_ind)):
                sup_scores = scores[ind,batch_sup[ind,:]]
                inf_scores = scores[ind,batch_inf[ind,:]]
                loss += torch.mean(torch.log(torch.sigmoid(sup_scores - inf_scores))) 
                #loss += torch.mean(sup_scores - inf_scores)
            loss /= len(batch_ind)  
            
            loss.backward()
            optimizer.step()
            avg_loss += loss.item() * len(batch_ind)
            total_num += len(batch_ind)
        acc = check_model(model,f,x_test_data,rank_test_data,False)
        
        
        
        print(ep,avg_loss/total_num,acc)
        if ep % 10 == 0:
            check_model(model,f,x_test_data,rank_test_data, plot)
            
        if acc > best:
            best = acc
            no_improve_ep = 0
            best_model = deepcopy(model.state_dict())
            
        if no_improve_ep > tol:
            break
            
    return best_model
```

# Hign dim function

```python
class QuadraticObj:
    
    def __init__(self, d=20, hidden_d=10):
        self.dim = d
        self.quadratic = np.random.randn(d,hidden_d)
        self.quadratic = np.matmul(self.quadratic, self.quadratic.T)
        self.linear = np.random.randn(d,1)
    
    def value_oracle(self, x):
        x_re = x.reshape(1,self.dim)
        obj = np.matmul(np.matmul(x_re,self.quadratic), x_re.T) + np.matmul(x_re, self.linear)
        return obj[0][0]
    
    def rank_oracle(self, x_list, top = -1):
        x_re_list = [x.reshape(1,self.dim) for x in x_list]
        value_list = [(np.matmul(np.matmul(x_re,self.quadratic), x_re.T) + np.matmul(x_re, self.linear))[0][0] for x_re in x_re_list]
        if top == -1:
            return np.argsort(value_list)
        else:
            return np.argsort(value_list)[:top]
        
    def gradient_oracle(self, x):   
        tensor_x = torch.tensor(x, requires_grad = True)
        x_re = tensor_x.reshape(1,self.dim)
        obj = torch.matmul(torch.matmul(x_re,torch.tensor(self.quadratic)), x_re.T) + torch.matmul(x_re, torch.tensor(self.linear))
        obj.backward()
        return tensor_x.grad.numpy()

    
    
class NonConvexObj:
    
    def __init__(self, d=20):
        self.dim = d
    
    def value_oracle(self, x):
        x_re = x.copy()
        obj = np.sum(100 * ((x_re[1:] - (x_re[:-1] ** 2))**2) + (1-x_re[:-1]) ** 2)
        return obj
    
    def rank_oracle(self, x_list, top = -1):
        x_re_list = [x.copy() for x in x_list]
        value_list = [np.sum(100 * ((x_re[1:] - (x_re[:-1] ** 2))**2) + (1-x_re[:-1]) ** 2) for x_re in x_re_list]
        if top == -1:
            return np.argsort(value_list)
        else:
            return np.argsort(value_list)[:top]
        
    def gradient_oracle(self, x):
        tensor_x = torch.tensor(x, requires_grad = True)
        obj = torch.sum(100 * ((tensor_x[1:] - (tensor_x[:-1] ** 2))**2) + (1-tensor_x[:-1]) ** 2)
        obj.backward()
        return tensor_x.grad.numpy()
```
